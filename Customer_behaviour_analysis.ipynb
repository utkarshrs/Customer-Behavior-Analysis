```python
import pandas as pd
```

```python
customers = pd.read_csv("customer_shopping_behavior.csv")
```

```python
customers.columns = customers.columns.str.lower()
customers.columns = customers.columns.str.replace(" ", "_")
customers = customers.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

### Handling Null values

##### Replacing null values of 'review_rating' attribute to median rating of each product catergory

```python
customers.isnull().sum()
```

```python
customers.category.unique()
```

```python
customers['review_rating'] = customers.groupby('category')['review_rating'].transform(lambda x: x.fillna(x.median()))
```

```python
customers.groupby('category')['review_rating'].median()
```

```python
customers.age.describe()
```

```python
# minimum age is 18 and maximum age is 70.
# dividing age to 4 age groups.

# Creating new attribute according to the age.
labels = ['Young Adult','Adult','Middle aged','Senior']
customers['age_group'] = pd.qcut(customers['age'], q=4, labels=labels) # qcut divides the data in 4 equal quartile bins.
```

```python
customers.columns
```

```python
customers['frequency_of_purchases'].unique()
```

```python
# Replacing the 'frequency_of_purchases' with no of days for better analysis.
frequency = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

customers['purchase_frequency_days'] = customers['frequency_of_purchases'].map(frequency)
```
